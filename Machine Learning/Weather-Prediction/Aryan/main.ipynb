```python
##NOTE: Can only be used if there is no error while getting the data,otherwisefor the exact co-ordinates,timestamp prediction is not posssible
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from pandas import Series
from statsmodels.tsa.arima_model import ARIMA
```

```python
# load dataset,store the variables
df = pd.read_csv('data.csv')
print(df.head())
print("-----------------------------------------")
tempMin=df['temperatureMin']
print(tempMin)
print("-----------------------------------------")
tempMax=df['temperatureMax']
print(tempMin)
```

```python
#Visualising the dataframe.
df.plot()
plt.show()
print(df)
```

```python
#Use the Arima Model to make the prediction for Min Temp
model = ARIMA(tempMin, order=(1,0,0))
model_fit = model.fit(disp=0)
forecast_min = model_fit.forecast(steps=1)[0]
print(forecast_min)
```

```python
#Use the Arima Model to make the prediction for Max Temp
model = ARIMA(tempMax, order=(1,0,0))
model_fit = model.fit(disp=0)
forecast_max = model_fit.forecast(steps=1)[0]
print(forecast_max)
```

```python
#Visualising the results obtained
df1=pd.DataFrame({"temperatureMin":forecast_min,"temperatureMax":forecast_max})
df=df.append(df1,ignore_index = True)
df.plot()
plt.show()
```
